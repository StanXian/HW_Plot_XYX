```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

```

```python
# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
city_df.head()
```

```python
ride_df = pd.read_csv(ride_data_to_load)
ride_df.head()
```

```python
# Combine the data into a single dataset
merge_table = pd.merge(city_df, ride_df, on="city")
# Display the data table for preview
merge_table.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
# Obtain the x and y coordinates for each of the three city types
gbc = merge_table.groupby("city")
ride = gbc["ride_id"].count()
fare_ave = gbc["fare"].mean()
driver_ave = gbc["driver_count"].mean()
city_type = city_df.set_index("city")["type"]

gbc_df = pd.DataFrame({
    "Number of Rides": ride,
    "Average Fare": fare_ave,
    "Number of Drivers": driver_ave,
    "Type of City": city_type
})
gbc_df
```

```python
# Build the scatter plots for each city types
rural = gbc_df[gbc_df['Type of City'] == 'Rural']
suburban = gbc_df[gbc_df['Type of City'] == 'Suburban']
urban = gbc_df[gbc_df['Type of City'] == 'Urban']

plt.scatter(rural['Number of Rides'], rural['Average Fare'], s = rural['Number of Drivers']*10, color = 'Yellow', edgecolor = 'black', label = 'Rural', alpha = .75)
plt.scatter(suburban['Number of Rides'], suburban['Average Fare'], s = suburban['Number of Drivers']*10, color = 'Blue', edgecolor = 'black', label = 'Suburban', alpha = .75)
plt.scatter(urban['Number of Rides'], urban['Average Fare'], s = urban['Number of Drivers']*10, color = 'Orange', edgecolor = 'black', label = 'Urban', alpha = .75)

# Create a legend
plt.legend(loc="upper right")

# Incorporate a text label regarding circle size
plt.title('Pyber Ride Sharing Data(2016)')
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare(S)")

# Save Figure
plt.savefig("Images/BubblePlot.png")

plt.grid()
plt.show()
```

```python
# Show plot
plt.show()
```

```python
gbt = merge_table.groupby('type')['type', 'fare', 'ride_id', 'driver_count']
gbt.head()
```

## Total Fares by City Type

```python
# Calculate Type Percents

gbt = merge_table.groupby('type')['type', 'fare', 'ride_id', 'driver_count']
total_fare = gbt.sum()['fare']

# Build Pie Chart
labels = ["Rural", "Suburban", "Urban"]
colors = ["Yellow", "Blue", "Orange"]

explode = (0 , 0, .3)
plt.pie(total_fare, startangle = 90, colors = colors, explode = explode, labels = labels, autopct = "%1.1f%%", shadow = True)

plt.title('% of Total Fares by City Type')
plt.axis("equal")
plt.show()

# Save Figure
plt.savefig("Images/total_fare.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_ride = gbt.count()['ride_id']

# Build Pie Chart
labels = ["Rural", "Suburban", "Urban"]
colors = ["Yellow", "Blue", "Orange"]

explode = (0 , 0, .3)
plt.pie(total_ride, startangle = 90, colors = colors, explode = explode, labels = labels, autopct = "%1.1f%%", shadow = True)

plt.title('% of Total Fares by City Type')
plt.axis("equal")
plt.show()

# Save Figure
plt.savefig("Images/total_ride.png")
```

```python
# Show Figure
plt.show()
```

```python
# Calculate Driver Percents
total_drivers = gbt.sum()['driver_count']


# Build Pie Charts
labels = ["Rural", "Suburban", "Urban"]
colors = ["Yellow", "Blue", "Orange"]

explode = (0 , 0, .3)
plt.pie(total_drivers, startangle = 90, colors = colors, explode = explode, labels = labels, autopct = "%1.1f%%", shadow = True)

plt.title('% of Total Fares by City Type')
plt.axis("equal")
plt.show()

# Save Figure
plt.savefig("Images/total_drivers.png")
```
